## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

```python
import numpy as np
```

```python
import pylab as plt
import seaborn as sns
from scipy.stats import expon
import matplotlib.pyplot as plt
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def generate_uniform_random_numbers(bottom, ceiling, count):
    x = uniform.rvs(size=count)
    randoms = bottom + (ceiling - bottom) * x
    return randoms

random_numbers_1 = generate_uniform_random_numbers(bottom=10, ceiling=15, count=100)

random_numbers_2 = generate_uniform_random_numbers(bottom=10, ceiling=60, count=1000)



```

```python
plt.figure(figsize=(10, 5))

# Histogram for random_numbers_1
plt.subplot(1, 2, 1)
plt.hist(random_numbers_1, bins=10, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution (bottom=10, ceiling=15, count=100)')

# Histogram for random_numbers_2
plt.subplot(1, 2, 2)
plt.hist(random_numbers_2, bins=10, color='green')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Uniform Distribution (bottom=10, ceiling=60, count=1000)')

plt.tight_layout()
plt.show()
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# your code here
def generate_normal_random_numbers(mean, std_dev, count):
    random_numbers = np.random.normal(mean, std_dev, count)
    return random_numbers


random_numbers_1 = generate_normal_random_numbers(mean=10, std_dev=1, count=1000)


random_numbers_2 = generate_normal_random_numbers(mean=10, std_dev=50, count=1000)
```

How are the two distributions different?

```python
# your answer below

plt.figure(figsize=(10, 5))

# Distribution for random_numbers_1
plt.subplot(1, 2, 1)
plt.hist(random_numbers_1, bins=30, color='blue')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Normal Distribution (mean=10, std_dev=1, count=1000)')

# Distribution for random_numbers_2
plt.subplot(1, 2, 2)
plt.hist(random_numbers_2, bins=30, color='green')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Normal Distribution (mean=10, std_dev=50, count=1000)')

plt.tight_layout()
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
def generate_exponential_random_numbers(mean, size):
    random_numbers = np.random.exponential(scale=mean, size=size)
    return random_numbers


random_numbers_1 = generate_exponential_random_numbers(mean=10, size=10)

random_numbers_2 = generate_exponential_random_numbers(mean=10, size=100)

# Plot the distributions
plt.figure(figsize=(10, 5))

# Distribution for random_numbers_1
plt.subplot(1, 2, 1)
plt.hist(random_numbers_1, bins=100, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Exponential Distribution (mean=10, size=10)')

# Distribution for random_numbers_2
plt.subplot(1, 2, 2)
plt.hist(random_numbers_2, bins=100, color='green', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Exponential Distribution (mean=10, size=100)')

plt.tight_layout()
plt.show()
```
